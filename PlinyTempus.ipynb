```python
plinius = ['\n'.join(f'<s>{s}</s>' for s in df['sentence'][i:i+30]) for i in range(0, len(df), 30)]
```

```python
p = """Given a list of 10 text blocks containing Latin sentences tagged with <s> tags, create a new list where each Latin sentence is replaced with its English translation tagged with <t> tags.

Rules:
1. Keep the same block structure
2. Replace each <s>Latin text</s> with <t>English translation</t>
3. Maintain the newline separators
4. Translate accurately and literally
5. Return only the translations without explanations

Process each sentence independently and preserve the exact format:
Input format:  <s>Latin sentence</s>
Output format: <t>English translation</t>"""

```

```python
from groq import Groq

def lat2eng(text: str) -> str:
    client = Groq()
    prompt = f"Translate this Latin text to English. ONLY translation: {text}"
    return client.chat.completions.create(
        model="llama3-groq-70b-8192-tool-use-preview",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.1,
        top_p=0.1,
        stream=False
    ).choices[0].message.content.strip()

# Test
print(lat2eng("et nos praeterierint"))
```

```python
translated = ['\n'.join(f'<t>{lat2eng(s[3:-4])}</t>' for s in block.split('\n')) for block in plinius]
```

```python
translated = []
for block in plinius:
    translations = '\n'.join(f'<t>{lat2eng(s[3:-4])}</t>' for s in block.split('\n'))
    translated.append(translations)
```
